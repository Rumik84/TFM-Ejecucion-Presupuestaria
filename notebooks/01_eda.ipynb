# EDA — Ejecución Presupuestaria TFM

Análisis exploratorio inicial del dataset curado por CCAA.

**Requisito:** haber ejecutado antes `python scripts/run_pipeline.py --only etl`.

```python
import pandas as pd
import plotly.express as px

from src.storage import SQLiteRepository

repo = SQLiteRepository()
facts = repo.load_ejecucion()
print(facts.shape)
facts.head()
```

## Ejecución acumulada por CCAA y fase

```python
agg = facts.groupby(['ccaa_slug', 'fase'])['importe_eur'].sum().reset_index()
px.bar(agg, x='ccaa_slug', y='importe_eur', color='fase', barmode='group')
```
